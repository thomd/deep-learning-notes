# SciKit Learn

```python
from sklearn.model_selection import train_test_split
from sklearn import datasets
import numpy as np
import pandas as pd
import seaborn as sns
```

## Iris Dataset KNN Classification

```python
from sklearn.neighbors import KNeighborsClassifier

X, y = datasets.load_iris(as_frame=True, return_X_y=True)
data, data['target'] = X, y
data
```

```python
sns.pairplot(data, hue='target');
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)

classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)

classifier.score(X_test, y_test)
```

## Synthetic Data NN Classification

```python
import numpy as np
X, y = datasets.make_classification(n_features=4, n_redundant=0, n_informative=4, n_classes=3)
```

```python
data = pd.DataFrame(np.c_[X, y])
data
```

```python
sns.pairplot(data, hue=4);
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)

classifier = KNeighborsClassifier()
classifier.fit(X_train, y_train)

classifier.score(X_test, y_test)
```
